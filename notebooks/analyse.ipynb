```python
def detect_outliers(value, upper, lower):
    if value >= upper:
        return 'upper'
    elif value <= lower:
        return 'lower'
    return 'normal'
    
outliers = pd.DataFrame(index=df.index)
```

```python
# Book_ratings_cuont
ratingCount_Q1 = df['Book_ratings_count'].quantile(0.25)
ratingCount_Q3 = df['Book_ratings_count'].quantile(0.75)

IQR = ratingCount_Q3 - ratingCount_Q1

upper = ratingCount_Q3 + (1.5 * IQR)
lower = ratingCount_Q1 - (1.5 * IQR) 


outliers["book_ratings_count"] = df["Book_ratings_count"].apply(lambda value: detect_outliers(value, upper, lower))
```

```python
# gross sales
grossSales_Q1 = df['gross sales'].quantile(0.25)
grossSales_Q3 = df['gross sales'].quantile(0.75)

IQR = grossSales_Q3 - grossSales_Q1

upper = grossSales_Q3 + (1.5 * IQR)
lower = grossSales_Q1 - (1.5 * IQR)

outliers["gross_sales"] = df["gross sales"].apply(lambda value: detect_outliers(value, upper, lower))
```

```python
# units sold
uniteSold_Q1 = df['units sold'].quantile(0.25)
uniteSold_Q3 = df['units sold'].quantile(0.75)

IQR = uniteSold_Q3 - uniteSold_Q1

upper = uniteSold_Q3 + (1.5 * IQR)
lower = uniteSold_Q1 - (1.5 * IQR) 

outliers["units_sold"] = df["units sold"].apply(lambda value: detect_outliers(value, upper, lower))
```

```python
# classic < 1800
# early modern 1800 - 1949
# modern 1950 - 1999
# contemprorary > 2000

dfCleaning['publishing_year'] = dfCleaning['publishing_year'].fillna(0)
year = dfCleaning['publishing_year']
dfCleaning['era'] = np.where(year == 0, 'Unknown', np.where(year < 1800, "classic", np.where(year <= 1949, 'Early Modern', np.where(year <= 1999, 'Modern', 'Contemprorary'))))
```

```python
dfCleaning.insert( 4, 'dialect', dfCleaning['language_code'].copy())
```

```python
dfCleaning.loc[
    (dfCleaning['language_code'] == 'en-US') | 
    (dfCleaning['language_code'] == 'en-GB') |
    (dfCleaning['language_code'] == 'en-CA') |
    (dfCleaning['language_code'] == 'eng'),
    'language_code'
    ] = 'English'

dfCleaning.loc[dfCleaning['language_code'] == 'spa', 'language_code'] = 'Spanish'
dfCleaning.loc[dfCleaning['language_code'] == 'fre', 'language_code'] = 'French'
dfCleaning.loc[dfCleaning['language_code'] == 'nl', 'language_code'] = 'Dutch'
dfCleaning.loc[dfCleaning['language_code'] == 'ara', 'language_code'] = 'Arabic'
dfCleaning['language_code'] = dfCleaning['language_code'].fillna("Undefined")
```
